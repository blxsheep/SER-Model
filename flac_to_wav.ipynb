```python
def flac_to_wav(read_path = '/content/drive/MyDrive/SER- Project/VISTEC_src/studio001/middle'): 
  paths = glob.glob(read_path + '/*') # buggs
  
  for path in paths: 
    print('read path = ' ,  path )
    filename = os.path.basename(path) 
    audio = AudioSegment.from_file(path)
    write_path = '/content/drive/MyDrive/SER- Project/src_as_wav/' + filename.replace('.flac','.wav')
    print('write path = ' , write_path)
    audio.export(write_path , format = 'wav')
    
  
```

```python
def list_src_dir() : 
  src_list = []
  
  for i in range(1,21):
    path  = '/content/drive/MyDrive/SER- Project/VISTEC_src/studio'
    if i < 10 :
      path += '00'
    else : path +='0'
    path+= str(i)
    clip =  path+ '/clip'
    middle = path +'/middle'
    src_list.append(clip)
    src_list.append(middle)
  return src_list 
```

```python
def add_white_noise(signal , noise_factor  = 0.1):
  noise = np.random.normal(0,signal.std(), signal.size)
  augmented_signal = signal + noise *noise_factor 
  return augmented_signal  
```
