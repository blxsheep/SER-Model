```python
def add_white_noise(signal , noise_factor  = 0.1):
  noise = np.random.normal(0,signal.std(), signal.size)
  augmented_signal = signal + noise *noise_factor 
  return augmented_signal  

```

```python
def augment_noise(import_path  , export_path): 
  signal , sr  =  librosa.load(import_path)
  for i in np.arange(0.01 , 0.11,0.01):
    augmented_signal  =  add_white_noise(signal , i)
    sf.write(export_path.replace('.wav','') + '_white_noise_'+str(i) + '.wav', augmented_signal , sr)
```

```python
def time_strectching(signal , stretch_rate): 
   return librosa.effects.time_stretch(signal,  stretch_rate)
```

```python
def augment_time_strecthing(import_path  , export_path): 
  signal , sr  =  librosa.load(import_path)
  for i in np.arange(0.1, 0.6,0.1):
    augmented_signal  =  time_strectching(signal , i)
    sf.write(export_path.replace('.wav','') + '_time_streching_'+str(i) + '.wav', augmented_signal , sr)
    
```

```python
src_path  =src
for filename  in os.listdir(src_path):
  print(filename)
  des_path =  des+filename 
  augment_noise(src_path + filename , des_path)
  augment_time_strecthing(src_path + filename , des_path)
```

```python

def random_gain(signal , min_gain_factor =2  , max_gain_factor =4 ):
  gain_factor =  random.uniform(min_gain_factor, max_gain_factor)
  return signal * gain_factor
```

```python
def augment_rand_gain(import_path  , export_path): 
  signal , sr  =  librosa.load(import_path)
  augmented_signal  =  random_gain(signal)
  sf.write(export_path.replace('.wav','') + '_random_gain' + '.wav', augmented_signal , sr)
  
```

```python
src_path  =src
for filename  in os.listdir(src_path):
  print(filename)
  des_path =  des+filename 
  augment_rand_gain(src_path + filename , des_path)

```
