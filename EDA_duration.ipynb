```python
def getDuration(path): 
  f = sf.SoundFile(path)
  return len(f)/f.samplerate
```

```python
src_path  ='/content/drive/MyDrive/SER- Project/src_as_wav/' 
sound_durations = []
for filename  in os.listdir(src_path):
  sound_durations.append(getDuration(src_path+filename))
```

```python
notify('Get sound duration : Done')
```

```python
print(sound_durations)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =  np.array(sound_durations)
q25, q75 = np.percentile(x, [25, 75])
bin_width = 2 * (q75 - q25) * len(x) ** (-1/3)
bins = round((x.max() - x.min()) / bin_width)
print("Freedman–Diaconis number of bins:", bins)
plt.hist(x, bins=bins);
plt.ylabel('Probability')
plt.xlabel('Sound Duration');
```
