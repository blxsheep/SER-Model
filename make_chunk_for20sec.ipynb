```python
from pydub import AudioSegment
from pydub.utils import make_chunks
src_path = '/content/drive/MyDrive/SER- Project/src_as_wav'
export_path =  '/content/drive/MyDrive/SER- Project/20sec_split_wav'
chunk_length_ms  = 20 *1000 
for filename  in os.listdir(src_path):
  file_path =  os.path.join(src_path,filename)
  org_audio = AudioSegment.from_file(file_path , "wav")
  chunks   = make_chunks(org_audio, chunk_length_ms)
  for i, chunk in enumerate(chunks):
    chunk_name = os.path.join(export_path,filename)
    print('Exporting : ' , chunk_name)
    chunk.export(chunk_name, format="wav")

```
